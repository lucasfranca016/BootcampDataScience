```python
#Desafio 2 - Aula 3: adicionar uma coluna mostrando a região de cada estado

lista = newdf.index
lista
```

```python
região = []

for i in lista:
    
    if i[0] == '1':
        
        região.append('Norte')
        
    if i[0] == '2':
        
        região.append('Nordeste')
    
    if i[0] == '3':
        
        região.append('Sudeste')
        
    if i[0] == '4':
        
        região.append('Sul')
        
    if i[0] == '5':
        
        região.append('Centro-Oeste')

newdf['Região Geográfica'] = região
```

```python
newdf
```

```python
#ordernando pelo gasto total:

newdf.sum(axis=1)

newdf['Total'] = newdf.sum(axis=1)

ordenadoPORtotal = newdf.sort_values('Total', ascending=False)

ordenadoPORtotal = ordenadoPORtotal.drop('Total', axis=1)

ordenadoPORtotal = ordenadoPORtotal.drop('Região Geográfica', axis=1)

ordenadoPORtotal.head(5).T.plot(figsize=(10,6))
```

```python
#Desafio 3 - Aula 3: Deixar o gráfico mais bonito

import matplotlib.ticker as ticker

ax = ordenadoPORtotal.head(5).T.plot(figsize=(10,6), cmap = cmaps['Miscellaneous'][15])

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.xlabel('Mês e ano')

plt.ylabel('Gasto com saúde pública (R$)')

plt.title('Gasto de cada estado com saúde pública por tempo')

plt.show()
```

```python
#Desafio 3 - Aula 5: Gasto por população, Rio de Janeiro e São Paulo:

popSP = 44 * pow(10, 6)

popRJ = 16.46 * pow(10, 6)

gastoRJ = ultimoMES['33 Rio de Janeiro']/popRJ

gastoSP = ultimoMES['35 São Paulo']/popSP

gastos = [gastoRJ, gastoSP]

rjsp = ['Rio de Janeiro', 'São Paulo']

plt.title('Gasto em saúde por habitante')

plt.xlabel('Estado')

plt.ylabel('Gasto (R$)')

plt.bar(rjsp, gastos , color='#184A45FF')

plt.show()
```

Como o gasto do último mês não está totalmente completo, o valor está relativamente baixo, na ordem de grandeza do número do número de habitantes, assim, temos aproximadamente 3 reais por habitante e quase 4 reais por habitante para RJ e SP respectivamente.
```python
obitos2 = obitos2.replace('-', 0, regex=True)
obitos2 = obitos2.astype(float)
obitos2.T[['Região Norte', 'Região Sul', 'Região Nordeste', 'Região Sudeste', 'Região Centro-Oeste']].plot()
```

Aqui eu peguei no site das tabelas a tabela referente aos óbitos por região federal e por ano/mês processamento. Eu excluí a coluna total e setei as região e estados como índices, análogo ao que foi feito em sala de aula. Em seguida, ao tentar transpor esse dataframe e plotar me deparei com o erro 'no numeric data found', então procurei pelos stackoverflow da vida e achei que poderia ser pelos dados estarem em números inteiros e não em float, então ao tentar transformar todos os dados em float foi retornado à mim que haviam strings nesse bolo de dados, aparentemente quando não há dados referentes a um mês de um certo estado eles botam um traço : '-' . Assim, eu substitui esses traços por zeros e depois transformei tudo em float e funcionou perfeitamente. Em seguida plotei um gráfico só de cada região como um todo, não dos estados em si. Em seguida farei um plot mais bonitinho.

```python
regioes = ['Região Norte', 'Região Sul', 'Região Nordeste', 'Região Sudeste', 'Região Centro-Oeste']

ax = obitos2.T[regioes].plot(figsize=(10,6))

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.ylabel('Número de mortos')

plt.xlabel('Ano/Mês')

plt.title('Gráfico de óbitos por região brasileira em função do tempo')

plt.show()
```

```python
#Aplicando um corte temporal para analisarmos a situação atual:

ax = obitos2.T[regioes][120:].plot(figsize=(10,6))

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.ylabel('Número de mortos')

plt.xlabel('Ano/Mês')

plt.title('Gráfico de óbitos por região brasileira em função do tempo')

plt.show()
```

Vemos que há um padrão parecido entre os anos de 2018 e 2019, onde há um pico de óbitos, melhor observado nas regiões sudeste, aproximadamente no mês de Setembro e outro por volta do início do ano. Ao analisarmos o ano de 2020 observamos um aumento substâncial de óbitos, provavelmente em grande parte por conta da pandemia que estamos enfrentando. Esse aumento pode observado em todas as regiões, contudo, de forma mais intensa na região Sudeste, provavelmente por ter as cidades com maior densidade populacional do país.
