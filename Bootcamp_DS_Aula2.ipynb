```python
#Quero pegar só os meses que possuem valor para todos os estados:

df.mean()

#O mean só pega as colunas que não tem valores vazios, fazendo esse trabalho para nós.
```

```python
colusavel = df.mean().index.tolist()
colusavel.insert(0,'Unidade da Federação')
colusavel
```

```python
#fazendo o recorte só dessas colunas:

newdf = df[colusavel]
newdf.head()
```

No nosso caso estamos eliminando essas colunas que tem valores vazios mas poderiamos adotar outra estratégia a depender do problema que estamos enfrentando. Outro tipo de solução poderia ser preencher esses valores com a média do estado em determinado ano, entre outras.

Fizemos uma alteração também transformando os index em lista para podermos alterar e com o .insert adicionamos à lista de index o string 'Unidade da federação', que corresponde ao nome da primeira coluna do DF, assim, nosso recorte incluirá essa coluna também.

```python
#Vamos agora fazer essa coluna Unidade da Federação virar o index:

newdf = newdf.set_index("Unidade da Federação")

newdf
```
