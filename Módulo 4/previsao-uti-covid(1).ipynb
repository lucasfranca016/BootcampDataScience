Os dos possuem o parâmetro Random_state, que é utilizado para passarmos uma seed para que o resultado seja sempre o mesmo, contudo, aleatório, assim, permitindo sua reprodutibilidade por outros cientistas.

Olhando bem para os dados vemos que o que pensavamos anteriormente ser o total de pacientes está errado, na coluna patient visit indetifier vemos que o mesmo paciente pode ser contado mais de uma vez, assim, precisamos fazer uma manipulação em nosso banco de dados.

```python
#colunas contínuas

feat_cont_col = dados.iloc[:, 13:-2].columns
feat_cont_col
```

```python
#Vamos começar preenchendo os valores vazios e usar só as colunas com valores contínuos

dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index = False)[
    feat_cont_col
].fillna(method='bfill').fillna(method='ffill')
```

```python
#Contudo vamos criar uma função que preenche só essas colunas contínuas:

def preencher(dados):
    feat_cont_col = dados.iloc[:, 13:-2].columns
    feat_cont = dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index = False)[feat_cont_col].fillna(method='bfill').fillna(method='ffill')
    feat_cat = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]
    dados_finais = pd.concat([feat_cat, feat_cont, saida], ignore_index=True, axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
```

```python
dados_limpos = preencher(dados)
dados_limpos.head()
```

```python
#Vemos que o count é 1925 em tudo mas alguns não pq são colunas vazia,então temos que limpar

dados_limpos.describe()
```

```python
dados_limpos = dados_limpos.dropna()
```

```python
#Vamos tirar o paciente q já entra indo pra UTI tb:

remover = dados_limpos.query("WINDOW == '0-2' and ICU==1")['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos = dados_limpos.query("PATIENT_VISIT_IDENTIFIER not in @remover")
```

```python
dados_limpos.describe()
```

```python
#Com essa função nós teremos uma linha para cada paciente, só os dados da primeira, e caso ele em
#algum momento tiver ido para a UTI sua primeira linha terá a coluna ICU com o valor 1.

def prepare(rows):
    if(np.any(rows['ICU'])):
        rows.loc[rows['WINDOW'] == '0-2', 'ICU'] = 1
    
    return rows.loc[rows['WINDOW'] == '0-2']

dados_limpos = dados_limpos.groupby("PATIENT_VISIT_IDENTIFIER").apply(prepare)
```

### Super desafio 6

```python
#Aqui só adicionei uma função que deixa de dar o fill back e after quando a linha tem ICU = 1.

def na(rows):
        rows.loc[rows["ICU"] != 1] = rows.loc[rows["ICU"] != 1].fillna(method='bfill').fillna(method='ffill')
        return rows
    
def preencher2(dados):
    feat_cont_col = dados.iloc[:, 13:-2].columns
    feat_cont = dados.groupby("PATIENT_VISIT_IDENTIFIER", as_index = False)[list(feat_cont_col) + ["ICU"]].apply(na)
    feat_cont = feat_cont.drop("ICU", axis=1)
    feat_cat = dados.iloc[:, :13]
    saida = dados.iloc[:, -2:]
    dados_finais = pd.concat([feat_cat, feat_cont, saida], ignore_index=True, axis=1)
    dados_finais.columns = dados.columns
    return dados_finais
```

```python
#Todo o processo de tratamento novamente, igual ao de antes

dados_limpos2 = preencher2(dados)
dados_limpos2 = dados_limpos2.dropna()
remover2 = dados_limpos2.query("WINDOW=='0-2' and ICU==1")['PATIENT_VISIT_IDENTIFIER'].values
dados_limpos2 = dados_limpos2.query("PATIENT_VISIT_IDENTIFIER not in @remover2")
dados_limpos2= dados_limpos2.groupby("PATIENT_VISIT_IDENTIFIER").apply(prepare)
```

```python
y = dados_limpos2['ICU']
x = dados_limpos2[xco].drop(['ICU'], axis=1)

np.random.seed(73246)

xtrain, xteste, ytrain, yteste = train_test_split(x, y, stratify=y)
```

```python
modelo3 = DummyClassifier()
modelo3.fit(xtrain, ytrain)

yprediction = modelo3.predict(xteste)
accuracy_score(yteste, yprediction)
```

```python
modelo = LogisticRegression()
modelo.fit(xtrain, ytrain)
yprediction = modelo.predict(xteste)
accuracy_score(yteste, yprediction)
```
