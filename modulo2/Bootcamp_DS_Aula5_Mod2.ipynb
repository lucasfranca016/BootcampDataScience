Vamos aplicar uma função para transformar isso em ano e mês númericos. Para isso usaremos o map.

```python
#Criando um dicionário para correlacionar cada mês ao seu respectivo número
meses = {
    "Jan": 1,
    "Fev": 2,
    "Mar": 3,
    "Abr": 4,
    "Mai": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Set": 9,
    "Out": 10,
    "Nov": 11,
    "Dez": 12,
}

#Criando uma função que lê a string contendo ano e mês e transforma em valores númericos para passarmos par ao date
def transfEMdate(anomes):
    ano = int(anomes[:4])
    mes = anomes[5:]
    mesnmr = meses[mes]
    return date(ano,mesnmr,1)

#Usando o map para usar a função em cada índice transformando em um date
mensal.index = mensal.index.map(transfEMdate)

mensal.head()
```

Agora o próprio python reconheci o index como uma data e o pandas já lida com esse formato de uma maneira diferente, por exemplo, tiramos o comando do locator, que separava nosso eixo x em valores de 12 em 12 e agora ele sozinho já interpreta que é melhor dispor os dados a cada 2 anos para não termos uma poluição visual!

Queremos agora plotar um gráfico com todos os estados juntos, contudo, teriamos várias colunas y(estados) e apenas uma x(data), o que nos geraria um problema, então, aprenderemos uma maneira de transpassar esse problema. Para isso faremos a data voltar a ser uma coluna e usaremos o método melt para "derreter" nosso dataframe de forma que os estados virem só mais uma coluna da nossa tabela.

```python
mensalABERTO = mensal.reset_index().melt(id_vars=['index'], value_vars=mensal.columns)
mensalABERTO.columns = ['AnoMesDia', 'uf', 'gasto']
mensalABERTO['AnoMesDia'] = mensalABERTO['AnoMesDia'].astype('datetime64')
mensalABERTO.head()
```
