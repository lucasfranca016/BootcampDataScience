```python
gastos = pd.read_csv('A165814189_28_143_208.csv', encoding='ISO-8859-1', skiprows = 3, sep = ';', skipfooter=12,
                thousands='.', decimal=',')

gastos.head()
```

```python
gastos = gastos.drop('Total',axis=1)
```

```python
vacinas = vacinas.drop('Total',axis=1)
vacinas = vacinas[:-1]
```

```python
colusavel = gastos.mean().index.tolist()
colusavel.insert(0,'Unidade da Federação')
gastos = gastos[colusavel]
gastos.head()
```

```python
gastos = gastos.set_index("Unidade da Federação")

gastos = gastos.T
```

```python
from datetime import date

meses = {
    "Jan": 1,
    "Fev": 2,
    "Mar": 3,
    "Abr": 4,
    "Mai": 5,
    "Jun": 6,
    "Jul": 7,
    "Ago": 8,
    "Set": 9,
    "Out": 10,
    "Nov": 11,
    "Dez": 12,
}

def transfEMdate(anomes):
    ano = int(anomes[:4])
    mes = anomes[5:]
    mesnmr = meses[mes]
    return date(ano,mesnmr,1)

gastos.index = gastos.index.map(transfEMdate)

gastos.head()
```

```python
gastosABERTO = gastos.reset_index().melt(id_vars=['index'], value_vars=gastos.columns)
gastosABERTO.columns = ['AnoMesDia', 'uf', 'gasto']
gastosABERTO['AnoMesDia'] = gastosABERTO['AnoMesDia'].astype('datetime64')
gastosABERTO.head()
```

```python
gastosABERTO['ano'] = gastosABERTO['AnoMesDia'].dt.year
gastoANO = gastosABERTO.groupby(['uf', 'ano']).sum().reset_index()
gastoANO.head()
```

```python
gastoANO['uf'] = gastoANO['uf'].str[3:]
gastoANO
```

```python
vacinas = vacinas.set_index("Região/Unidade da Federação")
vacinas = vacinas.T
```

```python
vacinasABERTO = vacinas.reset_index().melt(id_vars=['index'], value_vars=vacinas.columns)
vacinasABERTO.columns = ['ano', 'uf', 'vacinas']
vacinasABERTO
```

```python
reg = ['Região Norte', 'Região Sul', 'Região Nordeste', 'Região Centro-Oeste', 'Região Sudeste']
vacinasABERTO.query('uf != @reg')
vacinasABERTO2 = vacinasABERTO.query('uf != @reg')
```

```python
vacinasABERTO2['uf'] = vacinasABERTO2['uf'].str[2:]
display(vacinasABERTO2.head())
gastoANO.head()
```

```python
vacinasABERTO2['ano'] = vacinasABERTO2['ano'].astype(int)
```

```python
anos = []
for i in range(2008,2020):
    anos.append(i)
vac = vacinasABERTO2.query('ano in @anos')
gast = gastoANO.query('ano in @anos')
vac['uf'] = vac['uf'].str.strip()
```

```python
anoGASTO = gast.merge(vac)
su = ['Rio de Janeiro', 'São Paulo', 'Espírito Santo', 'Minas Gerais']
sud = anoGASTO.query('uf in @su')
sud['vacinas'] = sud['vacinas'].astype(int)
sud.head()
```

Tratamento:

    -Primeiramente importei as 2 bases de dados, a de quantidade de vacinas aplicadas por região e ano, e a de gasto em saúde pública por reginão e ano/mês, me atentando que apesar de serem csv estão separadas por ';' e que no arquivo tem algumas linhas que não pertecem à tabela, mas são linhas que descrevem o que está na tabela, um descrição sucinta, logo, pulo algumas linhas iniciais e finais, ondes os números de linhas puladas foram determinados abrindo o arquivo csv no bloco de notas.
    
    -Após isso, seleciono as linhas não vazias da tabela de gastos e transformo os anos/mês em formato date, trocando a sigla de cada mês por seu número correspondente e em seguida dou um melt nessa tabela para formatar de forma mais fácil para fazer o gráfico que almejo e cria uma coluna com os anos.
   
    -Faço um processo análogo de melt na tabela de vacinas e em ambas nomeio as colunas de forma semelhante. Assim, ajeito a coluna de Unidade da federação de ambas para que fiquem com strings iguais para cada nome de estado.
    
    -Retiro os dados de região da tabela de vacinas para que fiquem somente os estados, transformo o tipo do valor de anos da tabela vacinas em inteiro para que sejam interpretados como números.
    
    -Por fim, como ambas as tabelas abrangem intervalos de anos diferentes eu faço em cada uma o recorte dos anos em comum que ambas abrangem e junto ambas as tabelas em uma única tabela, orientando essa junção pela colunas de Unidade da federação, ou 'uf'.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker


plt.figure(figsize=(10,6))
ax = sns.lineplot(data=sud, x='ano', y='gasto', hue='uf')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
ax.xaxis.set_ticks(range(2008,2020,1))
plt.xlabel('Tempo (ano)')
plt.ylabel('Gasto com saúde pública (R$)')
plt.title('Gasto de cada estado do sudeste com saúde pública x tempo')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=sud, x='ano', y='vacinas', hue='uf')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.legend(bbox_to_anchor=(1,1), loc="upper left")
ax.xaxis.set_ticks(range(2008,2020,1))
plt.xlabel('Tempo (ano)')
plt.ylabel('Quantidade de vacinas aplicadas')
plt.title('Quantidade de vacinas aplicas em cada estado do sudeste x tempo')
plt.grid()
plt.show()
```

Análise do gráfico de vacinação:

    -Observamos um padrão desregular no gráfico e que intuitivamente nós leva a crer que em média a quantidade de vacinação no sudeste está diminuindo com o passar dos anos.

```python
s = []

for i in range(0,48):
    s.append('Sudeste')
    
sud['regiao'] = s

sud2  = sud.groupby(['regiao','ano']).sum()
sud2
```

Neste linha anterior estou adicionando uma coluna com o valor 'Sudeste' para todas as linhas com intuito de agrupar todas e fazer uma análise da região sudeste como um todo.

```python
import numpy as np
from matplotlib.font_manager import FontProperties

fontP = FontProperties()
fontP.set_size('xx-small')

plt.figure(figsize=(10,6))
ax = sns.scatterplot(data=sud2, x='gasto', y='vacinas', label='Pontos Vacinas x Gasto')
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))

plt.legend(bbox_to_anchor=(1,1), loc="upper left")
plt.xlabel('Gasto com saúde pública (R$)')
plt.ylabel('Quantidade de vacinas aplicadas')
plt.title('Quantidade de vacinas aplicadas na região sudeste por ano x Gasto da região sudeste com saúde pública por ano')
plt.grid()

m, b = np.polyfit(sud2['gasto'], sud2['vacinas'], 1)
bla = plt.plot(sud2['gasto'], m*sud2['gasto'] + b, label='Linear fit')

plt.legend( title='Legenda', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
print(f'O coeficiente angular do linear fit foi {m:.4f} e o coeficiente linear foi {b:.0f} .')
```

Para complementar a análise eu queria fazer um gráfico da quantidade de vacinas x gasto com saúde pública da região sudeste como um todo, assim, eu plotei os pontos gerados cruzando essas duas variáveis e fiz um "fit linear" desses pontos, ou seja, uma regressão linear, que é um método para criar a melhor reta que descreva o comportamento dos pontos apresentados, assim, o método chega nos melhores valores para o coeficiente angular da reta( o coeficiente que dá a inclinação da reta) e para o coeficiente linear da reta( o ponto em x onde a reta cruza o valor y=0) e plota um gráfico utilizando esses coeficientes os valores de x fornecidos (aqui, os valores de gasto ).

As primeiras conclusões que podemos obter deste gráfico são:

    -Observamos que quanto maior o investimento com saúde pública mais baixo a quantidade de vacinas aplicadas na região sudeste. Como sabemos que ambas as quantidades estão ordenados pelo passar do anos, temos implicitamente que com o passar dos anos o gasto com saúde pública vem aumentando enquanto a média do número de vacinas aplicadas vem diminuindo, resultado que é refletido nos gráficos posteriores por analisarem os mesmos índices.
    
    -Por conta do coeficiente angular ser muito baixo temos que a variação no eixo x é muito maior que a variação do eixo y, assim, podemos levantar a hipótese que o investimento em saúde na região Sudeste vem crescendo muito mais do que a média de vacinas aplicas na região vem diminuindo.
