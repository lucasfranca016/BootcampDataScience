A conclusão que temos aqui é que plotar um histograma para os mais de 3000 comspostos aqui é uma bela de uma porcaria, então iremos pegar os 5 mais recorrentes, aqueles com maior frequência, e faremos um histograma contendo somente esses 5, para isso, iremos fazer um recorte do nosso dataframe onde só apareçam esses 5 compostos.

```python
#Primeiro fazemos uma lista dos 5 com maior frequência
codcomp = dados['composto'].value_counts().index[:5]
codcomp
#Esse código vai na coluna composto, conta a ocorrência de cada composto único, pega os 5 primeiros e nos retorna
#uma lista somente com o nome deles.
```

```python
#Aplicando a filtragem, ressaltando aqui que o arroba se da porque codcomp foi definido fora dessa célula
dados.query('composto in @codcomp')
```

```python
#plotando agora uma gráfico com o data sendo essa filtragem que fizemos:

sns.set() #Para carregar as configs padrões do seaburn e deixar mais bonito

#Agora vamos usar o matplotlib para nomear os eixos e etc:
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
sns.countplot(x = 'composto', data = dados.query('composto in @codcomp'))
```

```python
#Desafio 1 -2 : ordernar o gráfico countplot

plt.figure(figsize=(10,8))
ax = sns.countplot(x = 'composto', data = dados.query('composto in @codcomp'), order = codcomp)
ax.set_title('Top 5 Compostos')
ax.set(xlabel = 'Nome do Composto', ylabel = 'Frequência do Composto')
plt.show()
```

Esse gráfico se chama box plot, sendo um gráfico que quanto mais pontos em um valor, mais pontinhos vai ter ali, com exceção da área onde se encontra a mediana, que é a barra azul, onde se encontram a maior parte dos valores e segundo a metodológia de criação do próprio gráfico, 50% de todos os pontos reside no intervalo de mais ou menos 25 % desse intervalo onde a mediana está inserido, ou seja, 50% dos valores está entre mais ou menos 0,5. Digo mais ou menos aqui no intuito de representar o intervalo de -0,5 até 0,5. Desafio 5:

Quando fazemos gráficos menores estamos suprimindo detalhes, como por exemplo, no gráfico countplot feito para as 5 drogas com maior uso, não conseguimos reconhecer visualmente a diferença entre algumas colunas, contudo, se aumentarmos o tamanho do gráfico podemos passar a perceber visualmente esse tipo de nuância pois os intervalos graficamente serão maiores, logo, mais perceptíveis. Outro tipo de manipulação é o tamnho dos bins do histograma, assim, podemos fazer os intervalos serem menos a fim de termos descrições mais detalhadas de nossa amostragem, logo, poderiamos ter o máximo de detalhes caso fizessemos intervalos infinitesimalmente pequeno, aproximando esse gráfico a uma curva, processo análogo a fazer uma integral.
