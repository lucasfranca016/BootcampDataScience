Ressaltamos aqui a relevância de se comparar variáveis, logo, se torna em evidência uma quantidade que diz justamente uma relação entre variáveis, a correlação.

```python
dados.loc[:,'g0':'g50'].corr()
```

Quanto mais próximo de 1 ou -1 mais correlacionada são as variáveis, quanto mais próximo de 0 menor correlacionadas estão. Agora vamos apresentar essas informações de uma maneira melhor, uma matriz de correlação.


```python
import numpy as np

# Compute the correlation matrix
corr = dados.loc[:,'g0':'g50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

 O código que pegamos veio com um parâmetro vmax = 0.3, limitando os valores que iria mostrar na direita, então o retiramos. 
 
Lembrando que: correlação não implica causalidade! As correlações podem indicar um possível caso que seja passível de uma examinação mais detalhada.


```python
#Vamos fazer isso agora para as células

# Compute the correlation matrix
corr_cel = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_cel, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_cel, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Gui comentou que para não cairmos em correlações que sejam coincidências há sites os quais são registrados os itens a serem testados, os parâmetros e a matemática a ser usada, assim, teremos certeza de que os pesquisadores que realizaram esse trabalho não estão manipulando a matemática usada, por exemplo, para chegar em um resultado tendencioso.

```python
#Desafio 5 Neste desafio usei conhecimento obtido dos códigos dos alunos Igor Mozetic e Luan Carlos Oliver
#https://github.com/luancsoliver/imersao.dados-alura/blob/main/Imersao_Dados_Aula_03.ipynb
#https://github.com/IgorMozetic/Imersoes_Alura/blob/main/Imers%C3%A3o_Dados/Aula3/Aula3_Imersao_Dados.ipynb

pri = dados.loc[:,'g0':'g50']
sec = dados.loc[:,'c0':'c50']
osdois = pri.join(sec)
osdois
```

Primeiro fizemos um corte das colunas de c0 até c50 e g0 até g50 e depois juntamos os dois "colando lado-a-lado" através do join, o que poderia ser feito usando o .concat com o parâmetro axis=1.

```python
var = osdois.corr()
var2 = var.loc[:'g50','c0':]
var2
```

Aqui fizemos a correlação da junção dos recortes anteriores e após isso usamos o .loc para fazer um recorte nas linhas de g0 até g50 e nas colunas de c0 até c50, produzindo assim a nossa matriz quadrada ideal para ser usada no plot de covariância.

```python

# Compute the correlation matrix
corr1 = var2

# Generate a mask for the upper triangle
#mask = np.triu(np.ones_like(corr1, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr1, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.xlabel('Culturas de células')
plt.ylabel('Genes')
plt.title('Correlação entre células e genes')
plt.show()
```

Aqui simplesmente utilizei a matriz produzida como fonte de informação para o plot e retirei a mask que apaga a diagonal superior do quadrado, se não ocorreria perda de informação, ao contrário do caso anterior onde tanto as linhas como colunas são de g0 até g50.
