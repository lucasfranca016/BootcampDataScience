Ressaltamos aqui a relevância de se comparar variáveis, logo, se torna em evidência uma quantidade que diz justamente uma relação entre variáveis, a correlação.

```python
dados.loc[:,'g0':'g50'].corr()
```

Quanto mais próximo de 1 ou -1 mais correlacionada são as variáveis, quanto mais próximo de 0 menor correlacionadas estão. Agora vamos apresentar essas informações de uma maneira melhor, uma matriz de correlação.


```python
import numpy as np

# Compute the correlation matrix
corr = dados.loc[:,'g0':'g50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

 O código que pegamos veio com um parâmetro vmax = 0.3, limitando os valores que iria mostrar na direita, então o retiramos. 
 
Lembrando que: correlação não implica causalidade! As correlações podem indicar um possível caso que seja passível de uma examinação mais detalhada.


```python
#Vamos fazer isso agora para as células

# Compute the correlation matrix
corr_cel = dados.loc[:,'c0':'c50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_cel, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_cel, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Gui comentou que para não cairmos em correlações que sejam coincidências há sites os quais são registrados os itens a serem testados, os parâmetros e a matemática a ser usada, assim, teremos certeza de que os pesquisadores que realizaram esse trabalho não estão manipulando a matemática usada, por exemplo, para chegar em um resultado tendencioso.

```python
#Desafio 5 Neste desafio usei conhecimento obtido dos códigos dos alunos Igor Mozetic e Luan Carlos Oliver
#https://github.com/luancsoliver/imersao.dados-alura/blob/main/Imersao_Dados_Aula_03.ipynb
#https://github.com/IgorMozetic/Imersoes_Alura/blob/main/Imers%C3%A3o_Dados/Aula3/Aula3_Imersao_Dados.ipynb

pri = dados.loc[:,'g0':'g50']
sec = dados.loc[:,'c0':'c50']
osdois = pri.join(sec)
osdois
```

Primeiro fizemos um corte das colunas de c0 até c50 e g0 até g50 e depois juntamos os dois "colando lado-a-lado" através do join, o que poderia ser feito usando o .concat com o parâmetro axis=1.

```python
var = osdois.corr()
var2 = var.loc[:'g50','c0':]
var2
```

Aqui fizemos a correlação da junção dos recortes anteriores e após isso usamos o .loc para fazer um recorte nas linhas de g0 até g50 e nas colunas de c0 até c50, produzindo assim a nossa matriz quadrada ideal para ser usada no plot de covariância.

```python

# Compute the correlation matrix
corr1 = var2

# Generate a mask for the upper triangle
#mask = np.triu(np.ones_like(corr1, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr1, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.xlabel('Culturas de células')
plt.ylabel('Genes')
plt.title('Correlação entre células e genes')
plt.show()
```

Aqui simplesmente utilizei a matriz produzida como fonte de informação para o plot e retirei a mask que apaga a diagonal superior do quadrado, se não ocorreria perda de informação, ao contrário do caso anterior onde tanto as linhas como colunas são de g0 até g50.

Def machine learning: é uma área que ensina o computador a fazer algo sem que ele seja programado para isso, ou seja, ensinamos por exemplo um computador a classificar uma imagem sem que a gente cria um conjunto de regras para ele fazer isso. Uma das maneiras de fazer isso é dar ao computador vários exemplos. Para construir modelos satisfatórios precisamos de muitos dados e poder computacional.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x= dados_combinados.select_dtypes('float64') #pega só os g e c

y= dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state=376)

modelo_rlogistica = LogisticRegression(max_iter=1000)

#treinar

modelo_rlogistica.fit(x_treino, y_treino)

#predição: vamos passar o x_teste e o y_teste, que são as respostas e usaremos um parâmetro score para ver o quão
#bem nosso modelo está prevendo os valores:

modelo_rlogistica.score(x_teste, y_teste)

#para sumir com o warning vamos setar um parâmetro de iterações máximas no nosso modelo.
```

O x são os dados utilizados para treinar o modelo, que no nosso caso vão ser os dados celulares e o y é o que queremos prever.

Para teste se funcionou precisamos de dados de compostos que não foram usados no treino para testá-los, assim, vamos dividir nossos dados, uma parte para treinar e outra para testar. Para isso usaremos o train test split do próprio scikit.

Será que 63% é um bom número de acertos? Vamos comparar com modelos mais simples, como por exemplo, comparar com um modelo que chuta que tudo é 1, e o própri scikit learning tem uma ferramenta para isso.

Vamos aumentar os níveis com um for e ir vendo o quão melhor vão ser esses resultados.

```python
from sklearn.tree import DecisionTreeClassifier

x= dados_combinados.select_dtypes('float64')

y= dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify = y, random_state=376)

test=[]
treino=[]

for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth= i )
    modelo_arvore.fit(x_treino,y_treino)
    test.append(modelo_arvore.score(x_teste,y_teste))
    treino.append(modelo_arvore.score(x_treino,y_treino))
    #estamos salvando o quão bem ele ta treinando e o quão bem estão sendo os resultados
```

```python
sns.lineplot(x=range(1,15), y = test, label='teste')
sns.lineplot(x=range(1,15), y = treino, label='treino')
```

Quanto mais profundidade na árvore de treino, melhor ele fica em adivinhar os próprios dados de treino, mas a partir de uma certa profundidade ele piora em acertar os dados de teste. Esse problema se chama overfitting. Precisamos de outras estratégias para melhorar esse resultado. Uma delas é criar várias árvores a partir de amostras aleatórias dos dados, o random forest.

Ao mudar a pergunta para se mais de uma MOA é ativado temos uma menor quantidade de respostas True, ou seja, temos menos candidatos que ativam mais de um MOA, assim, menos "caminhos" que levam à um True ou não, assim, não tornando só a previsão mais "fácil", mostrado pelo dummy ter 95% de precisão, mas também ao treinamento e previsão do algoritmo.

Em resumo, quando perguntamos se é x composto ou não temos uma maior facilidade de treinar um algoritimo para isso pois ele tem padrões mais claros.
